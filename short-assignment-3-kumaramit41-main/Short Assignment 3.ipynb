\begin{align*}
E_p(\mathbf{w},b) = -\sum_{m\in\mathcal{M}}(\mathbf{w}^T\mathbf{x}_n + b)^T t_n
\end{align*}

\begin{align*}
\mathbf{w}^{(t+1)} &\leftarrow & \mathbf{w}^{(t)} - \eta \frac{\partial E_p(\mathbf{w},b)}{\partial \mathbf{w}} = \mathbf{w}^{(t)} + \eta \mathbf{x}_n t_n\\
